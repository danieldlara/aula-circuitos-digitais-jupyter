```python
from BDD import *

x = BDD('x', ZERO, ONE)
y = BDD('y', ZERO, ONE)
z = BDD('z', ZERO, ONE)
```

```python
# Proposition 9
# x + x' = 1
# x . x' = 0

print("P9A", check(x.lor(x.negate()), ONE))
print("P9B", check(x.land(x.negate()), ZERO))

# Theorem 2-1
# x + 1 = 1

print(check(x.lor(ONE), ONE))

# Theorem 2-2
# x . 0 = 0

print(check(x.land(ZERO), ZERO))

# Theorem 3-1
# 0' = 1

print(check(ZERO.negate(), ONE))

# Theorem 4-1
# x + x = x
print(check(x.lor(x), x))
```

```python
# Theorem 5
print(check(x.negate().negate(), x))

# Theorem 6-1
print(check(x.lor(x.land(y)), x))

# Theorem 6-2
print(check(x.land(x.lor(y)), x))

```

```python
# Theorem 7-1

print(check(x.lor(x.negate().land(y)), x.lor(y)))

# Theorem 7-2

print(check(x.land(x.negate().lor(y)), x.land(y)))
```

```python
# Theorem 8-1

print(check(x.land(y.land(z)), (x.land(y)).land(z)))

# Theorem 8-2

print(check(x.lor(y.lor(z)), (x.lor(y)).lor(z)))
```

```python
# Theorem 9-1
# (x + y)' = x' . y'

print(check((x.lor(y)).negate(), (x.negate().land(y.negate())) ))

# Theorem 9-2
# (x . y)' = x' + y'

print(check((x.land(y)).negate(), (x.negate().lor(y.negate())) ))
```
